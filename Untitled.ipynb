```python
dropcolumns=fangzi['medv']
x=fangzi.drop(dropcolumns,axis=1)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
y=fangzi['medv']
SelectKBest = SelectKBest(f_regression,k=3)
bastFeature=SelectKBest.fit_transform(x,y)
SelectKBest.get_support()
x.columns[SelectKBest.get_support()]
```

```python
#特征归一化，此处是最大值最小值归一化:每个数据减去最小值再除以这列的最大值
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
for feature in fangzi.columns:
    fangzi['标准化'+feature]=scaler.fit_transform(fangzi[[feature]])
fangzi.head()
```

```python
from sklearn.model_selection import train_test_split#划分测试集和训练集
from sklearn.linear_model import LinearRegression#线性回归
from sklearn import metrics#指标
import warnings#忽略一些警告
warnings.filterwarnings('ignore')
```

```python
X=x
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred_class=model.predict(X_test)
y_pred_class
```

```python
#平均绝对误差
metrics.mean_absolute_error(y_test,y_pred_class)
```

```python
#均方误差：（每个值－估计值）的平方和的平均值
metrics.mean_squared_error(y_test,y_pred_class)
```

```python
#均方误差 竞赛常用
from math import sqrt
print(sqrt(metrics.mean_squared_error(y_test,y_pred_class)))
```

```python
#均方根对数误差
#对欠预测的惩罚更大，例如共享单车，欠预测意味着单车数量将不够
metrics.mean_squared_log_error(y_test,y_pred_class)
```

```python
#R平方
metrics.r2_score(y_test,y_pred_class)
```
